```python
import numpy as np
```

```python
columns = ['make', 'model', 'year', 'engine_hp', 'engine_cylinders', 'transmission_type', 'vehicle_style', 'highway_mpg', 'city_mpg', 'price']
df_linear = df_raw.loc[:,columns]
df_linear
```

```python
seed = 42
df_full_train_linear, df_test_linear = train_test_split(df_linear, test_size=0.2, random_state=seed)
df_train_linear, df_val_linear = train_test_split(df_full_train_linear, test_size=0.25, random_state=seed)
```

```python
len(df_train_linear), len(df_val_linear), len(df_test_linear)
```

```python
df_train_linear = df_train_linear.reset_index(drop=True)
df_val_linear = df_val_linear.reset_index(drop=True)
df_test_linear = df_test_linear.reset_index(drop=True)
```

```python
y_train_linear = np.log1p(df_train_linear.price.values)
y_val_linear = np.log1p(df_val_linear.price.values)
y_test_linear = np.log1p(df_test_linear.price.values)

del df_train_linear['price']
del df_val_linear['price']
del df_test_linear['price']
```

```python
df_train_linear
```

```python
def one_hot_encode_linear(columns):
    dv = DictVectorizer(sparse=True)

    train_dict = df_train_linear[columns].to_dict(orient='records')
    X_train = dv.fit_transform(train_dict)

    val_dict = df_val_linear[columns].to_dict(orient='records')
    X_val = dv.transform(val_dict)
    
    return X_train, X_val
```

```python
all_features = categorical + numerical
X_train_linear, X_val_linear = one_hot_encode_linear(all_features)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
```

```python
model = Ridge(solver='sag', alpha=1.0, random_state=42)
model.fit(X_train_linear, y_train_linear)
```

```python
y_pred_linear = model.predict(X_val_linear)
mean_squared_error(y_pred_linear, y_val_linear)
```

```python
alphas = [0, 0.01, 0.1, 1, 10]

rmses = []
for alpha in alphas:
    model = Ridge(solver='sag', alpha=alpha, random_state=42)
    model.fit(X_train_linear, y_train_linear)
    
    y_pred_linear = model.predict(X_val_linear)
    rmses.append(mean_squared_error(y_pred_linear, y_val_linear, squared=False))
```

```python
result = dict(zip(rmses, alphas))
result
```
